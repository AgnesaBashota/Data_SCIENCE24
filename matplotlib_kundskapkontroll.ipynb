### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

# Answer

###  Here's a short version explaining the two ways to plot with Matplotlib, and example code for both approaches:



```python
# Import necessary library
import matplotlib.pyplot as plt

# Plotting in Matplotlib can be done in two ways:
# 1. Object-Oriented API (Recommended)
# 2. Pyplot (State-Based) API

# Example of Object-Oriented API (recommended for complex plots)
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Object-Oriented API Plot')

# Example of Pyplot API (simpler, quick plots)
plt.figure()
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Pyplot API Plot')

# Show both plots
plt.show()
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

# Answer

### 1. Figure: The whole plot or window.
### 2. Axes: the actual plot area where data is drawn (can have multiple).
### 3. Axis: The X or Y axis showing the scale and labels.
### 4. Artist: Anything drawn on the plot (lines, text, lebels, etc)

### When plotting in Matplotlib, what is the expected input data type?

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1. Using a list
plt.figure()
plt.plot([1, 2, 3], [4, 5, 6], label='List Input')
plt.legend()
plt.title('Plot using List')

# 2. Using a NumPy array
x = np.array([1, 2, 3])
plt.figure()
plt.plot(x, x**2, label='NumPy Array Input')
plt.legend()
plt.title('Plot using NumPy Array')

# 3. Using a Pandas DataFrame
df = pd.DataFrame({'x': [1, 2, 3], 'y': [4, 5, 6]})
plt.figure()
plt.plot(df['x'], df['y'], label='Pandas DataFrame Input')
plt.legend()
plt.title('Plot using Pandas DataFrame')

# Show all plots
plt.show()
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(-4, 4, 100)  # 100 points between -4 and 4
y = x**2

# 1. Object-Oriented Approach
fig, ax = plt.subplots()  # Create a figure and an axis
ax.plot(x, y, label='y = x^2')  # Plot data
ax.set_title('Object-Oriented Approach')  # Set the title
ax.set_xlabel('x')  # Set the x-axis label
ax.set_ylabel('y')  # Set the y-axis label
ax.legend()  # Show the legend

# 2. Pyplot Approach
plt.figure()  # Create a new figure
plt.plot(x, y, label='y = x^2')  # Plot data
plt.title('Pyplot Approach')  # Set the title
plt.xlabel('x')  # Set the x-axis label
plt.ylabel('y')  # Set the y-axis label
plt.legend()  # Show the legend

# Show both plots
plt.show()
```
